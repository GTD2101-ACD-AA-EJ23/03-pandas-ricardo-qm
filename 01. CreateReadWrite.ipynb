En este JNB se revisará el tema de creación, lectura y escritura de datos en Pandas

# Introducción

Creación de colecciones básicas de datos con Pandas

```python
import pandas as pd
```

# Creación de datos

Los datos pueden venir de múltiples fuentes, desde datos internos al JNB hasta datos externos que provengan de bases de datos, archivos, etc. La estructura de datos estándar para trabajar en Pandas es el **Dataframe**.

```python
pd.DataFrame({'Yes': [50, 21], 'No': [131, 2]})
```

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 'Sue': ['Pretty good.', 'Bland.']})
```

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
```

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
```

# Series

Este es el otro tipo de datos que se maneja en Pandas. Una serie es una secuencia de valores indexados.

```python
pd.Series([1, 2, 3, 4, 5])
```

```python
pd.Series([30, 35, 40], index=['2015 Sales', '2016 Sales', '2017 Sales'], name='Product A')
```
