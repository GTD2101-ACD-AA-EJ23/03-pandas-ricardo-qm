# Combinación de dataframes

En ocasiones los datos pueden provenir de dos o más dataframes que deben ser combinados.

```python
canadian_youtube = pd.read_csv("https://media.githubusercontent.com/media/GTD2101-ACD-AA-EJ23/03-pandas-ricardo-qm/main/data/CAvideos.csv")
british_youtube = pd.read_csv("https://media.githubusercontent.com/media/GTD2101-ACD-AA-EJ23/03-pandas-ricardo-qm/main/data/GBvideos.csv")
```

```python
canadian_youtube
```

```python
british_youtube
```

```python

pd.concat([canadian_youtube, british_youtube])
```

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK')
```
